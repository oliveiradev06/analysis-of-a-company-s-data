```python
df.rename(columns={
    'ano': 'ano',
    'nivel_de_experiencia': 'senioridade',
    'tipo_de_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_do_empregado': 'residencia',
    'taxa_remota': 'remoto',
    'localizacao_da_empresa': 'localizacao',
    'tamanho_da_empresa': 'tamanho_empresa'
}, inplace=True)

display(df.columns)
```

```python
df["senioridade"].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

display(df['senioridade'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

display(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
})

display(df['tamanho_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

display(df['remoto'].value_counts())
```
